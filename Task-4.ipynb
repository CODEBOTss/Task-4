```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generating a random dataset
np.random.seed(0)
n_samples = 100
data = {
    'Age': np.random.randint(18, 70, size=n_samples),
    'Income': np.random.randint(20000, 100000, size=n_samples),
    'Score': np.random.randint(1, 100, size=n_samples)
}


df = pd.DataFrame(data)

summary = df.describe()
print("Summary Statistics:")
print(summary)


plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.histplot(df['Age'], bins=10, kde=True)
plt.title('Age Distribution')

plt.subplot(1, 3, 2)
sns.histplot(df['Income'], bins=10, kde=True)
plt.title('Income Distribution')

plt.subplot(1, 3, 3)
sns.histplot(df['Score'], bins=10, kde=True)
plt.title('Score Distribution')

plt.tight_layout()
plt.show()

# Visualizing outliers using box plots
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.boxplot(x=df['Age'])
plt.title('Age Box Plot')

plt.subplot(1, 3, 2)
sns.boxplot(x=df['Income'])
plt.title('Income Box Plot')

plt.subplot(1, 3, 3)
sns.boxplot(x=df['Score'])
plt.title('Score Box Plot')

plt.tight_layout()
plt.show()


correlation_matrix = df.corr()
print("Correlation Matrix:")
print(correlation_matrix)


sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
